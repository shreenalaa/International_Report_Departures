```python
def fill_missing_with_median_or_mean(data, feature, threshold):
    # Calculate the median and mean of the feature
    median_value = data[feature].median()
    mean_value = data[feature].mean()

    # Calculate the lower and upper bounds for identifying outliers
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    # Identify outliers in the feature
    outliers = (data[feature] < lower_bound) | (data[feature] > upper_bound)

    # Fill missing values based on whether there are outliers
    data.loc[outliers, feature] = median_value
    data.loc[~outliers, feature] = mean_value
```

```python
numerical_columns = df_Departures.select_dtypes(include=np.number).columns
for column in numerical_columns:
    if column not in ["Month","Year"]:
        threshold_for_outliers = 1.5 
        fill_missing_with_median_or_mean(df_Departures, column, threshold_for_outliers)
       
```

```python
numerical_columns = df_Passengers.select_dtypes(include=np.number).columns
for column in numerical_columns:
    if column not in ["Month","Year"]:
        threshold_for_outliers = 1.5 
        fill_missing_with_median_or_mean(df_Passengers, column, threshold_for_outliers)
```

```python
carrier_counts = df_Passengers['carrier'].value_counts()

# Set a threshold for carriers with counts less than 5 to be grouped into "Other"
threshold = 16660

# Identify carriers with counts less than the threshold
small_carriers = carrier_counts[carrier_counts < threshold]

# Assign the count of small carriers to the "Other" category
carrier_counts['Other'] = small_carriers.sum()

# Keep carriers with counts greater than or equal to the threshold
carrier_counts = carrier_counts[carrier_counts >= threshold]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(carrier_counts, labels=carrier_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Carriers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
